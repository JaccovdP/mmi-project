```python
# ALWAYS RUN THIS ONE FIRST
import pandas as pd
import random
import math
import numpy as np
import matplotlib.pyplot as plt
import json
df = pd.read_csv('results_fixed.csv')
baseline_types = ['Baseline_Haptic', 'Baseline_Audio']
types = ['Haptic', 'Audio']
df_baseline_combined = df[df['feedbackType'].isin(baseline_types)]
df_baseline_haptic = df[df['feedbackType'] == 'Baseline_Haptic']
df_baseline_audio = df[df['feedbackType'] == 'Baseline_Audio']
df_combined = df[df['feedbackType'].isin(types)]
df_haptic = df[df['feedbackType'] == 'Haptic']
df_audio = df[df['feedbackType'] == 'Audio']
```

```python
print("VISUAL BASELINE COMBINED")
print(df_baseline_haptic.describe())
print("========================================")
print("VISUAL BASELINE FOR HAPTIC TEST")
print(df_baseline_haptic.describe())
print("========================================")
print("VISUAL BASELINE FOR AUDIO TEST")
print(df_baseline_audio.describe())
print("========================================")
print("TESTS COMBINED")
print(df_combined.describe())
print("========================================")
print("HAPTIC TEST")
print(df_haptic.describe())
print("========================================")
print("AUDIO TEST")
print(df_audio.describe())
print("========================================")
```

```python
for x in range(0,len(df_combined)):
    target = df_combined.iloc[x]['target'].replace('\'', '"')
    target = json.loads(target)
    path = df_combined.iloc[x]['Path'].replace('\'', '"')
    path = json.loads(path)

    length_path = float(df_combined.iloc[x]['Distance'])
    optimal_distance = float(math.hypot(target['x'] - path[0]['x'], target['y'] - path[0]['y']))

    radius = float(df_combined.iloc[x]['radius'])
    accuracy = (optimal_distance - radius) / length_path

    X = np.array([])
    Y = np.array([])
    for i in path:
        X = np.append(X, i['x'])
        Y = np.append(Y, i['y'])
    X = np.append(X, target['x'])
    Y = np.append(Y, target['y'])

    calc_distance = 0
    for i in range(1, len(X) - 1):
        calc_distance += math.hypot(X[i] - X[i-1], Y[i] - Y[i-1])

    print(calc_distance)

    plt.scatter(X[-1],Y[-1], color="green", s=500)
    plt.scatter(X[0], Y[0], color="red", s=500)
    plt.plot(X, Y, color="black")
    print("Row number: %s | Participant: %s | Type: %s | Test nr: %s" % (x + 1, df_combined.iloc[x]['participant'], df_combined.iloc[x]['feedbackType'], df_combined.iloc[x]['test_nr']))
    print("Time: %s | Length of path: %s \nDistance between start and target: %s" % (df_combined.iloc[x]['Time'], length_path, optimal_distance - radius))
    print("Accuracy: %s" % accuracy)
    plt.savefig("results/" + str(x) + ".png")
    plt.show()
```

```python
accuracies_haptic = []
accuracies_audio = []
best_time = df_combined['Time'].min()
worst_time = df_combined['Time'].max()
print(best_time)
print(worst_time)
for x in range(0,len(df_haptic)):    
    target = df_haptic.iloc[x]['target'].replace('\'', '"')
    target = json.loads(target)
    path = df_haptic.iloc[x]['Path'].replace('\'', '"')
    path = json.loads(path)

    length_path = float(df_haptic.iloc[x]['Distance'])
    optimal_distance = float(math.hypot(target['x'] - path[0]['x'], target['y'] - path[0]['y']))

    radius = float(df_haptic.iloc[x]['radius'])
    if length_path == 0:
        accuracy = ( 1 + (best_time / float(df_haptic.iloc[x]['Time'])) ) / 2
    else:
        accuracy = ( ((optimal_distance - radius) / length_path) + (best_time / float(df_haptic.iloc[x]['Time'])) ) / 2
    
    accuracies_haptic.append(accuracy)
    
for x in range(0,len(df_audio)):    
    target = df_audio.iloc[x]['target'].replace('\'', '"')
    target = json.loads(target)
    path = df_audio.iloc[x]['Path'].replace('\'', '"')
    path = json.loads(path)

    length_path = float(df_audio.iloc[x]['Distance'])
    optimal_distance = float(math.hypot(target['x'] - path[0]['x'], target['y'] - path[0]['y']))

    radius = float(df_audio.iloc[x]['radius'])
    
    if length_path == 0:
        accuracy = ( 1 + (best_time / float(df_audio.iloc[x]['Time'])) ) / 2
    else:
        accuracy = ( ((optimal_distance - radius) / length_path) + (best_time / float(df_audio.iloc[x]['Time'])) ) / 2
    
    accuracies_audio.append(accuracy)


data = [accuracies_haptic, accuracies_audio]
fig = plt.figure(figsize=(15,10))
ax = fig.add_axes([0,0,1,1])
ax.boxplot(data)
ax.set_ylabel('Accuracy ( ((optimal distance - radius) / path length) + (best time / time) )')
ax.set_xlabel('Test')
ax.set_xticklabels(['Haptic', 'Audio'])
fig.savefig("results/combined_accuracy_boxplot.png", bbox_inches='tight', dpi=150)
```
